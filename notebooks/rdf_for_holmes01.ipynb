```python
from typing import TextIO
from rdflib import URIRef
import urllib.parse

def rename_str(x):
    x = str(x).replace('http://', 'http___')
    x = urllib.parse.quote(x, safe='/#_')
    return URIRef(x.replace('http___', 'http://'))

def holmes_watson_name_change(_s):
    if _s == "{}:{}".format(title_en, 'Holmes'):
        _s = 'AllTitle:Holmes'
    elif _s == "{}:{}".format(title_en, 'Watson'):
        _s = 'AllTitle:Watson'
    return _s


def ffunc10(title_en, g, f:TextIO, rdflib_dict, story_len):
    g10 = Graph()
    g10 += g.triples((None, None, None))
    g10 -= g.triples((None, rdflib_dict['kgc']['time'], None))
    g10 -= g.triples((None, rdflib_dict['kgc']['source'], None))
    g10 -= g.triples((None, rdflib_dict['rdfs']['label'], None))
    g10 -= g.triples((rdflib_dict[title_en]['metadata'], None, None))

    g_make = Graph()


    for i in range(1, story_len+1):
        iii = str(i).zfill(3)
        predicate_list = list(g10.objects(rdflib_dict[f'{title_en}'][iii], rdflib_dict['kgc']['hasPredicate']))
        subject_list = list(g10.objects(rdflib_dict[f'{title_en}'][iii], rdflib_dict['kgc']['subject']))
        if len(predicate_list)==1 and len(subject_list)==1:
            predicate=predicate_list[0]
            subject=subject_list[0]
            for _, _, o in g10.triples((rdflib_dict[f'{title_en}'][iii], None, None)):
                _s, _p, _o = map(rename_str, (subject,predicate,o))
                _s, _p, _o = map(lambda x: x.n3(g.namespace_manager), (_s, _p, _o))
                _s, _p, _o = map(urllib.parse.unquote, (_s, _p, _o))
                _s = holmes_watson_name_change(_s)
                _o = holmes_watson_name_change(_o)
                if '@ja' not in str(_o):
                    f.write("{}\t{}\t{}\n".format(_s, _p, _o))
                g_make+= g10.triples((subject, None, None))

    for s, p, o in g_make.triples((None, None, None)):
        _s, _p, _o = map(rename_str, (s, p, o))
        _s, _p, _o = map(lambda x: x.n3(g.namespace_manager), (_s, _p, _o))
        _s, _p, _o = map(urllib.parse.unquote, (_s, _p, _o))
        _s = holmes_watson_name_change(_s)
        _o = holmes_watson_name_change(_o)
        f.write("{}\t{}\t{}\n".format(_s, _p, _o))


def func03(title_en, story_len10, story_len09, story_len08):
    g = Graph()
    g.parse(f'{PROJECT_DIR}/../KGC_Data/KGRC-RDF/{title_en}.ttl')
    g.bind('wordInfo:predicate', rdflib.Namespace(f"http://kgc.knowledge-graph.jp/data/predicate/"))
    g.bind(f'{title_en}', rdflib.Namespace(f"http://kgc.knowledge-graph.jp/data/{title_en}/"))
    rdflib_dict = {key: rdflib.Namespace(value) for key, value in g.namespace_manager.namespaces()}
    with open(f'{PROJECT_DIR}/data/external/KGCdata/{title_en}/all.txt', 'w') as f:
        ffunc10(title_en, g, f, rdflib_dict, story_len10)

import os
for title_ja, (title_en, story_len10, story_len09, story_len08) in title_info_dict.items():
    print(title_ja)
    os.makedirs(f'{PROJECT_DIR}/data/external/KGCdata/{title_en}/', exist_ok=True, )
    func03(title_en, story_len10, story_len09, story_len08) if not title_ja == '踊る人形' else None

```
